```python
import json
import operator
import pandas as pd

with open("/content/merged_dataset.json") as outfile:
    data = json.load(outfile)
    raw_data = list(map(operator.itemgetter('raw_data'), data)) # this is when memory starts running out for me

df = pd.json_normalize(raw_data)
df.drop(df.columns.difference(['full_text']), 1, inplace=True)
print(df)

df.to_csv(r'/content/merged_dataset.csv')
```
